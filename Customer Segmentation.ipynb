```python
from datetime import datetime

# Converting 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Calculating Recency (Days since last purchase)
latest_date = df['Date'].max()
df['Recency'] = (latest_date - df['Date']).dt.days

# Calculating Frequency (Total Transactions per Customer)
frequency_df = df.groupby('Customer_ID')['Transaction_ID'].count().reset_index()
frequency_df.columns = ['Customer_ID', 'Frequency']

# Calculating Monetary Value (Total Amount Spent per Customer)
monetary_df = df.groupby('Customer_ID')['Total_Spent'].sum().reset_index()
monetary_df.columns = ['Customer_ID', 'Monetary']

# Merging RFM features into one dataset
rfm_df = df[['Customer_ID', 'Recency']].drop_duplicates().merge(frequency_df, on='Customer_ID').merge(monetary_df, on='Customer_ID')


# Fill any NaN values after conversion (if needed)
df.fillna(0, inplace=True)

# Display the RFM DataFrame
print(rfm_df.head())
```

```python
print(df.columns)
```

```python
df_frequency = df.groupby("Customer_ID")["Transaction_ID"].count().reset_index()
df_frequency.columns = ["Customer_ID", "Frequency"]
```

```python
print(df_frequency.columns)
```

```python
df_monetary = df.groupby("Customer_ID")["Total_Spent"].sum().reset_index()
df_monetary.columns = ["Customer_ID", "Monetary"]
```

```python
df_monetary.columns
```

```python
rfm_df = df[["Customer_ID", "Recency"]].drop_duplicates()
rfm_df = rfm_df.merge(df_frequency, on="Customer_ID", how="left")
rfm_df = rfm_df.merge(df_monetary, on="Customer_ID", how="left")

# Fill missing values (if any)
rfm_df.fillna(0, inplace=True)

# Display final RFM dataset
print(rfm_df.head())
```

```python
from sklearn.cluster import KMeans

# Select only RFM features
rfm_features = rfm_df[['Recency', 'Frequency', 'Monetary']]

# Standardize the data
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_features)

# Convert back to DataFrame
rfm_scaled_df = pd.DataFrame(rfm_scaled, columns=['Recency', 'Frequency', 'Monetary'])
```

```python
# Finding the optimal number of clusters using Elbow Method
wcss = []  # Within-cluster sum of squares
for k in range(1, 11):  # Testing k from 1 to 10
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(rfm_scaled_df)
    wcss.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(8,5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

silhouette_scores = []
for k in range(2, 11):  # Start from 2 clusters
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(rfm_scaled_df)
    score = silhouette_score(rfm_scaled_df, labels)
    silhouette_scores.append(score)

# Plot Silhouette Scores
plt.figure(figsize=(8,5))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score for Optimal K')
plt.show()
```

```python
# Select only RFM features
rfm_features = rfm_df[['Recency', 'Frequency', 'Monetary']]

# Standardize the data
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm_features)

# Convert back to DataFrame
rfm_scaled_df = pd.DataFrame(rfm_scaled, columns=['Recency', 'Frequency', 'Monetary'])
```

```python
# Apply K-Means with K=3
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
rfm_df['Cluster'] = kmeans.fit_predict(rfm_scaled_df)
```

```python
# Count of customers in each cluster
print(rfm_df['Cluster'].value_counts())

# Cluster-wise mean values of RFM features
cluster_analysis = rfm_df.groupby('Cluster')[['Recency', 'Frequency', 'Monetary']].mean()
print(cluster_analysis)
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='Cluster', y='Monetary', data=rfm_df)
plt.title('Customer Segmentation Based on Monetary Value')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=rfm_df['Recency'], y=rfm_df['Monetary'], hue=rfm_df['Cluster'], palette='viridis')
plt.title('Customer Segmentation (Recency vs. Monetary)')
plt.show()
```
